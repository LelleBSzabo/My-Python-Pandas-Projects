```python
# Average mark across all students

df["mark"].mean()
```

```python
# Average mark across male and average mark across female students

df.groupby("gender")["mark"].mean()
```

```python
# Max and min mark

max_mark = df["mark"].max()
min_mark = df["mark"].min()

print("Maximum mark:", max_mark)
print("Minimum mark:", min_mark)
```

```python
# Max mark for male and max mark for female students

max_mark = df.groupby("gender")["mark"].max()
min_mark = df.groupby("gender")["mark"].min()

print("Maximum mark:", max_mark)
print("Minimum mark:", min_mark)
```

```python
# Average mark for class 'Five'

df[df["class"]=="Five"]["mark"].mean()
```

```python
# Average mark for class 'Six' and male students only

df[(df["class"] == "Six") & (df["gender"] == "male")]["mark"].mean()
```

#Sort Values

```python
df["mark"].sort_values().head()
```

```python
df.loc[0:4, ["name", "mark"]].sort_values(by="mark", ascending=False)
```

```python
# find male student with their mark higher than 70 and sort by their class

df.loc[(df["gender"]=="male") & (df["mark"]>70)].sort_values(by="class")
```

```python
df.loc[(df["gender"]=="female") & (df["mark"]>70)]
```

```python
#Write code to find information of female students who are above the classroom average. Sort the results in descending order.

df.loc[(df["gender"]=="female") & (df["mark"]>(df["mark"].mean()))].sort_values(by="mark", ascending=False)
```

```python
df.loc[(df["gender"]=="female") & (df["mark"]>df["mark"].mean())].sort_values(by="mark", ascending=False)
```
